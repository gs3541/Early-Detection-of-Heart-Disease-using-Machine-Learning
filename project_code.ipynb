```python
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
```

# Data Loading

```python
df= pd.read_csv("heart.csv")
df.head()
df_0=df.copy()
```

```python
df.info()
```

# Data Wrangling

# Checking for null values

```python
df.isnull().sum()
```

Since we don't have any null values, we can proceed.

## Encoding categorical variables

```python
df.dtypes
```

```python
print(df['Sex'].unique())
#Error 1
#For Sex feature Assigning M value as 0 anf F as 1
df.Sex = df.Sex.replace({"M":0,"F":1 })
df.head()
```

```python
print(df['ExerciseAngina'].unique())
df.ExerciseAngina = df.ExerciseAngina.replace({"N":0,"Y":1 })

df.head()
```

```python
# Chest Pain Type
print(df['ChestPainType'].unique())
df.ChestPainType = df.ChestPainType.replace({"ATA":0,"NAP":1, "ASY":2, "TA":3, })
df.head()
```

```python
#Resting ECG
print(df["RestingECG"].unique())
df.RestingECG = df.RestingECG.replace({"Normal":0,"ST":1, "LVH":2})
df.head()
```

```python
# ST_Slope
print(df["ST_Slope"].unique())
df.ST_Slope = df.ST_Slope.replace({"Up":0,"Flat":1, "Down":2})
df.head()
```

```python
df.dtypes
```

## Correlation: HeatMap

```python

features=dict(df.dtypes)
df.corr()
```

```python
#HeatMap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
plt.savefig('heatmap.png')
plt.show()
```

```python
# We checked the correlation values between fields to check if any value was greater than 0.7
# If yes, then those one of those fields can be chosen to drop
import numpy as np
cor_matrix = df.corr()
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
# print(upper_tri)
to_drop = [column for column in upper_tri.columns if any(abs(upper_tri[column]) > 0.7)]

print("Columns to drop:", to_drop)
print("Columns to keep", set(df.columns) - set(to_drop))
```

## Violin and Swarmplot Hybrid for RestingECG vs Age for Heart Diseases cases or Normal cases : This will help us see count and distribution of different RestingECG categories across Age values while using different colors depending on Heart Diseases present or not.


```python
plt.figure(figsize=(15, 8))

def add_cosmetics(title="graph", 
                  xlabel="x",ylabel="y"):
    plt.title(title, fontsize=28)
    plt.xlabel(xlabel, fontsize=20)
    plt.ylabel(ylabel, fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    sns.despine()
add_cosmetics()

sns.violinplot(y=df_0["RestingECG"], x=df_0["Age"], data=df_0, color='yellow', cut=0)
sns.swarmplot(y=df_0["RestingECG"], x=df_0["Age"], data=df_0, color='blue')
add_cosmetics("RestingECG vs Age","Age","RestingECG")
plt.savefig("divergence_age_resting_ecg.png")
```

```python
# Find outliers in the dataset.
from scipy import stats
from scipy.stats import zscore

df_temp = df
df_no_outliers = df
for feature in columns:
  sns.boxplot(df_temp[feature])
  print(df_temp.shape)
  plt.show()
  z_scores = stats.zscore(df_temp[feature])

  abs_z_scores = np.abs(z_scores)
  filtered_entries = (abs_z_scores > -3)
  df_no_outliers = df_temp[filtered_entries]
  filtered_entries = (abs_z_scores < 3)
  df_no_outliers = df_temp[filtered_entries]

  df_temp = df_no_outliers
  df_no_outliers = []
  

print(df.shape)
print(df_temp.shape)
```

# Dataset Split

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

from IPython.core.pylabtools import figsize
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
print(len(df))
print(len(df_temp))
Y = df['HeartDisease'].to_numpy()
X = df.drop(['HeartDisease'], axis=1)
Y_wo = df_temp['HeartDisease'].to_numpy()
X_wo = df_temp.drop(['HeartDisease'], axis=1)

#min-max normalization
scaler = MinMaxScaler()

#Convert the data to numpy array
X= scaler.fit_transform(X)
X_wo= scaler.fit_transform(X_wo)

# 301
# Prepare testing and training data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, shuffle=True, random_state=6)
X_train_wo, X_test_wo, Y_train_wo, Y_test_wo = train_test_split(X_wo, Y_wo, test_size=0.2, shuffle=True, random_state=6)
print(X_train.shape, Y_train.shape)
print(X_train_wo.shape, Y_train_wo.shape)

```

# Machine Learning Classifiers

## Logistic Regression

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, plot_confusion_matrix, plot_roc_curve

clf = LogisticRegression(random_state=0).fit(X_train, Y_train)
clf_wo = LogisticRegression(random_state=0).fit(X_train_wo, Y_train_wo)
Y_pred = clf.predict(X_test)
Y_pred_wo = clf.predict(X_test_wo)
print("Confusion Matrix\n", confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
print()

figsize(6,8)
plot_roc_curve(clf, X_test, Y_test)
plt.savefig("Logistic_Pregression_AUROC.png")
plt.show()

print("\nLogistic Regression without Outliers\n")
print("Confusion Matrix\n", confusion_matrix(Y_test_wo, Y_pred_wo))
print(classification_report(Y_test_wo, Y_pred_wo))
print()

figsize(6,8)
plot_roc_curve(clf_wo, X_test_wo, Y_test_wo)
plt.savefig("Logistic_Pregression_AUROC_withoutOutliers.png")
plt.show()
```

## Decision Tree

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, roc_auc_score, plot_confusion_matrix, plot_roc_curve, accuracy_score

clf = DecisionTreeClassifier(random_state=0).fit(X_train, Y_train)
Y_pred = clf.predict(X_test)

clf_wo = DecisionTreeClassifier(random_state=0).fit(X_train_wo, Y_train_wo)
Y_pred_wo = clf.predict(X_test_wo)


print("Confusion Matrix\n", confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
print()

figsize(6,8)
plot_roc_curve(clf, X_test, Y_test)
plt.savefig("Decision_Tree_AUROC.png")
plt.show()


print("\n \n Decision Tree without Outliers\n")
print("Confusion Matrix\n", confusion_matrix(Y_test_wo, Y_pred_wo))
print(classification_report(Y_test_wo, Y_pred_wo))
print()

figsize(6,8)
plot_roc_curve(clf_wo, X_test_wo, Y_test_wo)
plt.savefig("Decision_Tree_AUROC_withoutOutliers.png")
plt.show()
```
